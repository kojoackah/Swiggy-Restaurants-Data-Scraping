```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.swiggy.com'
response = requests.get(url)

urls = []

soup = BeautifulSoup(response.content,"html.parser")
for a_href in soup.find_all("a",href =True):
    urls.append(a_href["href"])
```
